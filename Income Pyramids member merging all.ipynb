```python
def ms_change(row):
    colnames = ['16_ms_x', '16_ms_y', '17_ms', '18_ms', '19_ms',
       '20_ms', '21_ms', '22_ms', '23_ms', '24_ms_x', '24_ms_y']
    for col in colnames:
        if row['16_ms_x'] != row[col]:
            return True
    return False
```

```python
df_bride['ms_change'] = df_bride.apply(ms_change,axis=1)
```
